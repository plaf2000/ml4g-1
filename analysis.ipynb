```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_DATA_PATH = "Data/processed/data_train.npz"
```

```python
training_data = np.load(TRAIN_DATA_PATH)
X1: np.ndarray = training_data["X1"]
X2: np.ndarray = training_data["X2"]

y_X1: np.ndarray = training_data["labels_X1"]
y_X2: np.ndarray = training_data["labels_X2"]


X = np.concatenate((X1, X2), axis=0)
y = np.concatenate((y_X1, y_X2), axis=0)
print(np.mean(X[:, :, 0, 1], axis=0))

labels = [
    "DNase",
    "H3K4me1",
    "H3K4me2",
    "H3K4me3",
    "H3K27ac",
    "H3K27me3",
    "H3K36me3",
]

# plt.hist(X[:, 0, 0, 1], density=False, bins=100, histtype='bar', label=labels[0])
plt.hist(X[:, :, 0, 1], density=True, bins=20, histtype='bar', label=labels, range=(-1e4, 1e4))
plt.ylabel("Density")
plt.xlabel("Minimal distance of peak from TSS start")
plt.legend()
plt.show()
```

```python
plt.scatter(X[:, 0, 0, 1], y, s=X[:, 0, 0, 0], c=X[:, 0, 1, 1])
plt.xlim(-2e4, 2e4)
plt.show()
```
